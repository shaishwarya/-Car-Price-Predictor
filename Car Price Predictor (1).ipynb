```python
x=ndf.drop(["price","fueltype","enginetype","enginelocation"], axis=1)
x.head()
```

```python
y=ndf['price']
y[:5]
```

```python
from sklearn import preprocessing
cname=preprocessing.LabelEncoder()
dnumber=preprocessing.LabelEncoder()
cbody=preprocessing.LabelEncoder()
cnumber=preprocessing.LabelEncoder()
```

```python
x['CarName']=cname.fit_transform(x['CarName'])
x.head()
```

```python
x['CarName']=cname.fit_transform(x['CarName'])
x['doornumber']=dnumber.fit_transform(x['doornumber'])
x['carbody']=cbody.fit_transform(x['carbody'])
x['cylindernumber']=cnumber.fit_transform(x['cylindernumber'])
```

```python
from sklearn import linear_model
model=linear_model.LinearRegression()
model.fit(x,y)
model.score(x,y)
```
